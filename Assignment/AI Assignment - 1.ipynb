```python
#dd.isnull().sum().sum()
## Filling the Null Values

dd1 = dd.fillna(value=0)#can any values 
dd1
```

```python
#fill the Null values with the previous value
dd3=dd.fillna(method='pad')
dd3
```

```python
#fill the Null values with the next value
dd4=dd.fillna(method='bfill')
dd4
```

```python
#fill the Null values with the column before value
dd5=dd.fillna(method='pad', axis = 1)
dd5
```

```python
#fill the Null values with the next column  value
dd6=dd.fillna(method='bfill', axis=1)
dd6
```

```python
# insert the new element in the Null from the column head
dd7 = dd.fillna({'Age':'Wrong Age',
                'Salary':'No Salary'})
dd7
```

```python
#Eliminate the Nan from the Table
dd8 = dd.dropna()
dd8
```

```python
dd8 = dd.dropna(how='any')
dd8
```
